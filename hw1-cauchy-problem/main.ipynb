```python
import matplotlib.pyplot as plt
import data
import euler_method
import runge_kutta
import adams
from mpl_toolkits.mplot3d import Axes3D
```

```python
ans = euler_method.explicit_euler_method(data.lorenz, 0.01, 10000, [0, 1, 1.05], 28)
```

```python
assert len(ans) == 3
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs=ans[0], ys=ans[1], zs=ans[2])
plt.show()

```
