```python
import matplotlib.pyplot as plt
import data
import euler_method
import runge_kutta
import adams
import random
import lib
from mpl_toolkits.mplot3d import Axes3D
```

```python
ans = euler_method.explicit_euler_method(data.lorenz, 0.01, 10000, [0, 1, 1.05], 0.3)
```

```python
assert len(ans) == 3
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs=ans[0], ys=ans[1], zs=ans[2])
plt.show()

```

```python
assert len(ans) == 3
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs=ans[0], ys=ans[1], zs=ans[2])
plt.show()
```

```python
ans1 = euler_method.explicit_euler_method(data.lorenz, 0.001, 100000, [0, 1, 1.05], 10)
ans2 = euler_method.not_explicit_euler_method(data.lorenz, 0.001, 100000, [0, 1, 1.05], 10)
ans3 = runge_kutta.runge_kutta(data.lorenz, 0.001, 100000, [0, 1, 1.05], 10)
ans4 = adams.adams_method(data.lorenz, 0.001, 100000, [0, 1, 1.05], 10)
print(ans4[0][-1], ans4[1][-1], ans4[2][-1])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(xs=ans1[0], ys=ans1[1], zs=ans1[2], label='explicit euler')
ax.plot(xs=ans2[0], ys=ans2[1], zs=ans2[2], label='not explicit')
ax.plot(xs=ans3[0], ys=ans3[1], zs=ans3[2], label='runge-kutta')
ax.plot(xs=ans4[0], ys=ans4[1], zs=ans4[2], label='adams')
ax.legend()
plt.show()
```
